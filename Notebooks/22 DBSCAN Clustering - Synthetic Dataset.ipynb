# DBSCAN Clustering - Implementation on a Synthetic Dataset

## About

Goal of this kernel is to implement **DBSCAN** (Density-based spatial clustering of applications with noise) Clustering. We will be using a synthetic dataset derived from the make_moons function in sci-kit learn datasets module.

We will be:

* Creating a synthetic dataset using the sci-kit learn datasets module.
* Scaling the data using StandardScaler
* Using DBSCAN to identify clusters in the dataset.
* Visualizing the data before and after clustering.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
```

# Importing the dataset

```python
X, y = make_moons(n_samples=1000, noise=0.05, random_state=42)
```

```python
X[:5]
```

Dataset has 2 features.

# Scaling the data

```python
scaler = StandardScaler()
X = scaler.fit_transform(X);
```

```python
X.std(), X.mean()
```

# Visualizing data before Clustering

```python
plt.scatter(x=X[:, 0], y=X[:, 1]);
```

# Applying DBSCAN Clustering

```python
dbscan = DBSCAN(eps=0.3, min_samples=5)
```

```python
dbscan.fit(X)
```

```python
labels = dbscan.labels_
```

```python
np.unique(labels)
```

DBSCAN has found 2 clusters labelled '0' and '1'.

# Visualizing data after clustering

```python
plt.scatter(x=X[:, 0], y=X[:, 1], c=labels);
```

DBSCAN has correctly clustered the datapoints using the density of the points; this wouldn't have been possible with other algorithms like k-means, which would have divided the data into 2 clusters along the vertical axis.
